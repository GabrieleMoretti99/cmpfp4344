```python
import numpy
import math
import matplotlib.pyplot as plt

lp=[]
for n in range (0,10000):
  a=numpy.random.choice([-1,1],10000)
  #print(a)
  c=numpy.cumsum(a) #random walk di ogni walker
  #print(c)
  #plt.hist(c)
  #plt.show()
  lp.append(c[9])  #array con posizioni finali
print(lp)
plt.hist(lp)
max_pos=max(lp) #massima distanza dall'origine
print(max_pos)
n_max_pos=list(lp).index(max_pos) #pos. max distanza
print(n_max_pos)
```
